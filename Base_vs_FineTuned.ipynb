# **Convert Labels (0-4) to sentiments**
- Rating 3-4 will be positive -> 0
- Rating 0-1 will be negative -> 1
- Rating 2 will be neutral and ignored -> -1


***Datasets has labels (0-4 ratings) and texts***

```python
# View first 5 rows of the training split
for i in range(5):
    print(dataset["train"][i])
```

```python
def label_to_sentiment(example):
    if example["labels"] >= 3:
        return {"label": 1}
    elif example["labels"] <= 1:
        return {"label": 0}
    else:
        return {"label": -1}  # neutral, ignore

# Apply transformation
labeled_dataset = dataset["train"].map(label_to_sentiment)

# Filter out neutral reviews
labeled_dataset = labeled_dataset.filter(lambda x: x["label"] != -1)
```

# **Tokenizing the Dataset**
- Shuffling the dataset so it's randomized
- Split 80% for training and 20% for testing

Load the Tokenizer for DistillBERT to turn each label, review texts to make the model understand.


```python
from transformers import AutoTokenizer

split_dataset = labeled_dataset.shuffle(seed=42).train_test_split(test_size=0.2)

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize(example):
    return tokenizer(example["text"], truncation=True, padding="max_length")

tokenized_dataset = split_dataset.map(tokenize, batched=True)
```

# **Fine-tuning DistillBERT with PEFT**
**PEFT** = Parameter-Efficient Fine-Tuning
Basically instead of updating all of a large model’s weights (which is slow, memory-heavy, and expensive), PEFT lets you **freeze** most of the model and add a few small layers (adapters) that are trained. This is good because you get nearly the same accuracy but with way less training time.


**LoRA** = Low-Rank Adaptation.

It’s one of the most popular PEFT methods. Instead of updating a big weight matrix W inside the model (millions of numbers). LoRA says:
"Hey — let’s freeze W, and just add two tiny trainable matrices (A and B) that plug into W during training." So we only learn A and B, which are much smaller (like 8×64 instead of 768×768), much faster to train and can be added on top of existing models.

**Analogy:**
- Fine-tuning the whole model is like rewriting an entire book.

- Using LoRA is like just adding sticky notes to the important pages.

Using LoRA was especially useful for my dataset because I was working with a relatively small, specific set of Amazon shoe reviews. Fine-tuning the full DistilBERT model would have been slow, memory-intensive, and more prone to overfitting given the limited data. By using LoRA, I was able to fine-tune just a small part of the model (about 1% of the parameters), which made training much faster and more efficient. It also let me adapt the model to better understand the language and sentiment patterns specific to shoe reviews — such as complaints about sizing or praise for comfort — without needing huge computational resources. This made LoRA the perfect fit for building a lightweight, task-specific model using my own dataset.


*Takes two hours to train with T4 GPU*

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from peft import LoraConfig, get_peft_model, TaskType

# Step 1: Load base DistilBERT model for binary classification
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)

# Step 2: Add LoRA adapter using PEFT
peft_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,  # sequence classification task
    inference_mode=False,        # we're training, not just running inference
    r=8,                         # rank of adapter matrices
    lora_alpha=32,               # scaling factor
    lora_dropout=0.1,             # regularization
    target_modules=["q_lin", "v_lin"]  # ← this fixes the value error for DistilBERT
)


model = get_peft_model(model, peft_config)

# (Optional) Print # of trainable parameters to show LoRA is active
model.print_trainable_parameters()

# Step 3: Set training arguments
training_args = TrainingArguments(
    output_dir="./results",
    #evaluation_strategy="epoch",     # evaluate after every epoch
    logging_steps=10,
    num_train_epochs=3,              # train for 3 passes over the data
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    save_strategy="no",
    report_to="none"                 # disable wandb/logging
)

# Step 4: Create the Trainer object
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer=tokenizer
)

# Step 5: Train the model
trainer.train()
```

# **Evaluation and Comparison**

```python
metrics = trainer.evaluate()
print(metrics)
```

```python
import re
from transformers import pipeline, AutoModelForSequenceClassification

# Set up pipelines
fine_tuned_pipe = pipeline("text-classification", model=model, tokenizer=tokenizer)
base_model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
base_pipe = pipeline("text-classification", model=base_model, tokenizer=tokenizer)

# Keyword-based justification
positive_keywords = ["love", "fit", "perfect", "comfortable", "amazing", "beautiful", "great", "nice", "cute"]
negative_keywords = ["hurt", "uncomfortable", "bad", "tight", "blisters", "return", "disappointed", "pain", "cheap", "mid"]

def get_justification(text, label):
    reasons = []
    for word in positive_keywords:
        if re.search(rf"\b{word}\b", text.lower()):
            reasons.append(f"positive keyword: '{word}'")
    for word in negative_keywords:
        if re.search(rf"\b{word}\b", text.lower()):
            reasons.append(f"negative keyword: '{word}'")
    if not reasons:
        reasons.append("no strong sentiment keywords found")

    if label == "LABEL_1":
        sentiment = "Positive (LABEL_1)"
        filtered = [r for r in reasons if "positive" in r or "no strong" in r]
    else:
        sentiment = "Negative (LABEL_0)"
        filtered = [r for r in reasons if "negative" in r or "no strong" in r]

    return sentiment, "; ".join(filtered)

# Input loop
while True:
    review = input("\nEnter a shoe review (or type 'exit' to quit): ")
    if review.lower() == "exit":
        break

    # Fine-tuned model prediction
    fine_result = fine_tuned_pipe(review)[0]
    fine_sentiment, fine_reason = get_justification(review, fine_result["label"])

    # Base model prediction
    base_result = base_pipe(review)[0]
    base_sentiment, base_reason = get_justification(review, base_result["label"])

    # 🔍 Print results
    print("\nFine-Tuned Model")
    print(f"Prediction: {fine_sentiment}")
    print(f"Confidence: {round(fine_result['score'] * 100, 2)}%")
    print(f"Justification: {fine_reason}")

    print("\nPretrained Base Model")
    print(f"Prediction: {base_sentiment}")
    print(f"Confidence: {round(base_result['score'] * 100, 2)}%")
    print(f"Justification: {base_reason}")
```

```python
from transformers import pipeline
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Use your fine-tuned model
sentiment_pipe = pipeline("text-classification", model=model, tokenizer=tokenizer)

def classify_review(review):
    result = sentiment_pipe(review)[0]
    label = "positive" if result["label"] == "LABEL_1" else "negative"
    confidence = round(result["score"] * 100, 2)
    return label, confidence
```

```python
prompt = PromptTemplate.from_template(
    """You are an avid shoe-shopper and you are looking at all shoe reviews in Amazon. A shoe review was classified as {label} with {confidence}% confidence.
The review is: "{review}"

Explain why the label makes sense based on the wording of the review."""
)

llm = ChatOpenAI(model_name="gpt-3.5-turbo")
chain = LLMChain(llm=llm, prompt=prompt)
```

```python
while True:
    user_input = input("\nEnter a shoe review (or 'exit' to quit): ")
    if user_input.lower() == "exit":
        break

    label, confidence = classify_review(user_input)
    explanation = chain.run(label=label, confidence=confidence, review=user_input)

    print(f"\n🧠 Model prediction: {label.capitalize()} ({confidence}%)")
    print("🤖 Explanation from GPT:")
    print(explanation)
```
